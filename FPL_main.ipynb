```python
# Dependencies

import pandas as pd
from fpl import formation, plot_players_img
```

```python
# Reading dataset

players = pd.read_csv('fpl_data_processed.csv')
image_base = 'https://resources.premierleague.com/premierleague/photos/players/110x140/p'
base_url = 'https://fantasy.premierleague.com/api/'
```

# Data Analysis

### 1. Transfer In-Out

```python
# Top In

feaures_req = ['full_name', 'now_cost', 'photo', 'singular_name', 'influence', 'creativity',
           'threat', 'ict_index', 'minutes',
           'points_per_game', 'selected_by_percent', 'status',
           'club_name', 'next_1', 'next_2', 'next_3']

output = players.sort_values(by = 'transfers_in_event',
                    axis = 0,
                    ascending  = False).head(10)[feaures_req].reset_index(drop =True)

plot_players_img(output, image_base).plot()
output
```

```python
# Top Out

output = players.sort_values(by = 'transfers_out_event',
                    axis = 0,
                    ascending  = False).head(10)[feaures_req].reset_index(drop =True)

plot_players_img(output, image_base).plot()
output
```

### 2. Most Selected

```python
players.selected_by_percent = players.selected_by_percent.astype('float64')
players.sort_values('selected_by_percent', ascending=False)[feaures_req].head(10).reset_index(drop=True)
```

### 3. Top 11 Formation
(Without 3 players per team rule and without upper buy cap)

```python
# Defining best formation based on points

team_3_4_3_1 = formation(players, 3,4,3,1).players()
team_3_3_4_1 = formation(players, 3,3,4,1).players()
team_2_5_3_1 = formation(players, 2,5,3,1).players()
team_2_3_5_1 = formation(players, 2,3,5,1).players()
team_2_4_4_1 = formation(players, 2,4,4,1).players()
team_1_4_5_1 = formation(players, 1,4,5,1).players()
team_1_5_4_1 = formation(players, 1,5,4,1).players()


team_3_4_3_1_avg = team_3_4_3_1.points_per_game.mean()
team_3_3_4_1_avg = team_3_3_4_1.points_per_game.mean()
team_2_5_3_1_avg = team_2_5_3_1.points_per_game.mean()
team_2_3_5_1_avg = team_2_3_5_1.points_per_game.mean()
team_2_4_4_1_avg = team_2_4_4_1.points_per_game.mean()
team_1_4_5_1_avg = team_1_4_5_1.points_per_game.mean()
team_1_5_4_1_avg = team_1_5_4_1.points_per_game.mean()

req_dict = {'team_3_4_3_1':team_3_4_3_1_avg, 'team_3_3_4_1':team_3_3_4_1_avg, 
            'team_2_5_3_1':team_2_5_3_1_avg, 'team_2_3_5_1':team_2_3_5_1_avg, 'team_2_4_4_1':team_2_4_4_1_avg, 
            'team_1_4_5_1':team_1_4_5_1_avg, 'team_1_5_4_1':team_1_5_4_1_avg}

max_key = max(req_dict, key=req_dict.get)
print('Best 11 formation:', max_key)

forwd = int(max_key.split('_')[1])
midfl = int(max_key.split('_')[2])
defed = int(max_key.split('_')[3])
```

```python
# Plotting best formation

plot_players_img(formation(players, forwd,midfl,defed,1).players(), image_base).plot()
```
