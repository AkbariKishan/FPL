```python
# Dependencies

import pandas as pd
from fpl import formation, plot_players_img
```

```python
# Reading dataset

players = pd.read_csv('fpl_data_processed.csv')
image_base = 'https://resources.premierleague.com/premierleague/photos/players/110x140/p'
base_url = 'https://fantasy.premierleague.com/api/'
```

# Data Analysis

### 1. Transfer In-Out

Let's checkout players transfers statistics across the globe for this weekend's fixtures.

```python
# Top In: Top most trasferred-in players across the globe by FPL managers

feaures_req = ['full_name', 'now_cost', 'photo', 'singular_name', 'influence', 'creativity',
           'threat', 'ict_index', 'minutes',
           'points_per_game', 'selected_by_percent', 'status',
           'club_name', 'next_1', 'next_2', 'next_3']

output = players.sort_values(by = 'transfers_in_event',
                    axis = 0,
                    ascending  = False).head(10)[feaures_req].reset_index(drop =True)

plot_players_img(output, image_base).plot()
output
```

The above mentioned players are hot favorites by FPL managers. The reason can be players form, easier upcoming fixtures, lower cost or sometimes can be cheapest idle bench. 

```python
# Top Out: Top most trasferred-out players across the globe by FPL managers

output = players.sort_values(by = 'transfers_out_event',
                    axis = 0,
                    ascending  = False).head(10)[feaures_req].reset_index(drop =True)

plot_players_img(output, image_base).plot()
output
```

The above mentioned players are most transferred out by FPL managers across the globe. Here, the reason behind their relatively less popularity can be degrading form, difficult fixtures, red card, injury or 3 players per EPL team constraint. 

### 2. Most Selected

```python
# Most selected players: Players present with the most FPL teams

players.selected_by_percent = players.selected_by_percent.astype('float64')
output = players.sort_values('selected_by_percent', ascending=False)[feaures_req].head(10).reset_index(drop=True)

plot_players_img(output, image_base).plot()
output
```

The above list of players are present in most of the FPL manager's team. Here most import features are consistent delivery and cost.

### 3. ICT Index (Influence - Creativity - Threat)

`ICT` is an important parameter for FPL managers to quantify player's performance.

1. `Influence`:
Influence evaluates the degree to which a player has made an impact on a single match or throughout the season.

2. `Creativity`:
Creativity assesses player performance in terms of producing goalscoring opportunities for others. It can be used as a guide to identify the players most likely to supply assists.

3. `Threat`:
This is a value that examines a player's threat on goal. It gauges the individuals most likely to score goals.

```python
# Top players by ICT

players.ict_index = players.ict_index.astype('float64')
output = players.sort_values('ict_index', ascending=False)[feaures_req].head(10).reset_index(drop=True)

plot_players_img(output, image_base).plot()
output
```

The above plot indicates top players by ICT index. In general forwards, midfielders and attacking left/right backs have better ICT index due to their capability to create goal chances in the game.

### 4. Top Formation
(Without 3 players per team rule and without upper buy cap)

```python
# Defining best formation based on points

team_3_4_3_1 = formation(players, 3,4,3,1).players()
team_3_3_4_1 = formation(players, 3,3,4,1).players()
team_2_5_3_1 = formation(players, 2,5,3,1).players()
team_2_3_5_1 = formation(players, 2,3,5,1).players()
team_2_4_4_1 = formation(players, 2,4,4,1).players()
team_1_4_5_1 = formation(players, 1,4,5,1).players()
team_1_5_4_1 = formation(players, 1,5,4,1).players()


team_3_4_3_1_avg = team_3_4_3_1.points_per_game.mean()
team_3_3_4_1_avg = team_3_3_4_1.points_per_game.mean()
team_2_5_3_1_avg = team_2_5_3_1.points_per_game.mean()
team_2_3_5_1_avg = team_2_3_5_1.points_per_game.mean()
team_2_4_4_1_avg = team_2_4_4_1.points_per_game.mean()
team_1_4_5_1_avg = team_1_4_5_1.points_per_game.mean()
team_1_5_4_1_avg = team_1_5_4_1.points_per_game.mean()

req_dict = {'team_3_4_3_1':team_3_4_3_1_avg, 'team_3_3_4_1':team_3_3_4_1_avg, 
            'team_2_5_3_1':team_2_5_3_1_avg, 'team_2_3_5_1':team_2_3_5_1_avg, 'team_2_4_4_1':team_2_4_4_1_avg, 
            'team_1_4_5_1':team_1_4_5_1_avg, 'team_1_5_4_1':team_1_5_4_1_avg}

max_key = max(req_dict, key=req_dict.get)
print('Best 11 formation:', max_key)

forwd = int(max_key.split('_')[1])
midfl = int(max_key.split('_')[2])
defed = int(max_key.split('_')[3])
```

The above formation can be interpreted as,

`team_X-Y-Z-W`: 

`X`: No. of `forwards` in playing 11

`Y`: No. of `mid fielders` in playing 11

`Z`: No. of `defenders` in playing 11

`W`: `Goalkeeper` (default 1)

Above calculation indicates the formation with corresponding players, which results in maximum value of: Average_game_point/fixture_difficulty

Thumb rule is to select players based on parameters,
- Average points per game (indicating form)
- fixture difficulty (indicating opponent team + Home/Away parameter)

```python
# Plotting the best formation

plot_players_img(formation(players, forwd,midfl,defed,1).players(), image_base).plot()
```

Sometimes the above team can be biased to a a few clubs. The reason can be easier fixture and multiples players in the team performing exceptionally well.
